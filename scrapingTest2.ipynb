```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import *

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time

PATH = 'C:\Program Files (x86)\chromedriver.exe'
Lazada_url = 'https://www.lazada.com.my'
search_item = 'dyson vacuum' 
```

```python
# Select custom Chrome options
# to block unncesseary ads etc
options = webdriver.ChromeOptions()
# options.add_argument('--headless') 

# options.add_argument("--incognito")
options.add_argument('start-maximized') 
options.add_argument('disable-infobars')
options.add_argument('--disable-extensions')
```

```python
# open the chrome browser
driver = webdriver.Chrome(PATH,options=options)
driver.get(Lazada_url)
print(driver.title)


# look at html inspect. find search button element
search = driver.find_element_by_id('q')
search.send_keys(search_item)
search.send_keys(Keys.RETURN)

#find the infobox class
item_in_details = driver.find_elements_by_class_name("c5TXIP")


page1 = driver.find_element_by_class_name("cpF1IH")
page2 = page1.find_element_by_tag_name("li")
page3 = page2.get_property("title")

print(item_in_details[0].text)
print(page3)

```

```python
################--------------------MAIN LOOP------------------------------------###################

def get_item_prices(item):
    
    
    PATH = 'C:\Program Files (x86)\chromedriver.exe'
    Lazada_url = 'https://www.lazada.com.my'

    # Select custom Chrome options
    # to block unncesseary ads etc
    options = webdriver.ChromeOptions()
    options.add_argument("--incognito")
    options.add_argument('start-maximized') 
    options.add_argument('disable-infobars')
    options.add_argument('--disable-extensions')

    all_details=[]
    link_details=[]
    
    
    condition=True
    while condition:


        # open the chrome browser
        driver = webdriver.Chrome(PATH,options=options)
        driver.get(Lazada_url)

        # look at html inspect. find search button element
        search = driver.find_element_by_id('q')
        search.send_keys(item)
        search.send_keys(Keys.RETURN)



        time.sleep(5)
        #infobox
        item_in_details = driver.find_elements_by_class_name("c5TXIP")

        for item in item_in_details:

            #get the price and title
            item_titles = item.find_element_by_class_name('c16H9d').text
            item_prices = item.find_element_by_class_name('c13VH6').text

            #find the url href of the item
            p1 = item.find_element_by_tag_name("div")
            p2 = p1.find_element_by_tag_name("a")
            p3 = p2.get_property("href")

            #find the review and location
            rating1 = item.find_element_by_class_name('c15YQ9').text  


            #append inside a dictionary
            tempDict = {'title':item_titles,
                        'price':item_prices,
                        'review':rating1,
                        'link':p3}

            all_details.append(tempDict)
            link_details.append(p3)


        #navigate to next page
        try:
            driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[4]/div/ul/li[9]/a').click()

        except:
            condition=False
```
